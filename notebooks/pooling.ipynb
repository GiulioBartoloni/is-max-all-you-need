```python
class TinyModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.w = torch.nn.Parameter(torch.tensor(2.0))

    def forward(self, x):
        return (x * self.w).sum()

model = TinyModel()
print(model.w)   

x = torch.tensor([1., 2., 3.])
y = model(x)
print(y)

y.backward()
print(model.w.grad)
```

```python
class NewTinyModel(torch.nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x, mask):
        unsqueezed_mask = mask.unsqueeze(-1)

        masked = x * unsqueezed_mask
        final = masked.sum(dim=1)
        return final
    
model = NewTinyModel()

test = torch.ones(2,5,100)
mask = torch.tensor([[1,1,1,0,0],[1,1,0,0,0]])

y = model(test,mask)
print(y.shape)
print(y)
```
